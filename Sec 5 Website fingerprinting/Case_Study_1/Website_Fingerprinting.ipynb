```python
import numpy as np
import pandas as pd
import os
import glob
from collections import Counter
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
from itertools import combinations
from scipy.stats import ttest_ind
from tqdm import tqdm
import joblib
```

```python
data_folder = 'Traces/'

# Load CSV files
csv_files = glob.glob(os.path.join(data_folder, '*_power_traces.csv'))
print("Found CSV files:", len(csv_files))

features = []
labels = []

print("Reading CSV files and extracting features...")
```

```python
for csv_file in csv_files:
    df = pd.read_csv(csv_file)

    # Extract website name from filename
    website_name = os.path.basename(csv_file).split('_power_traces.csv')[0]

    # Use trace values (flattened) as features; exclude non-trace columns
    trace_cols = [col for col in df.columns if col.startswith("Trace_")]

    # Transpose: each trace is a feature vector
    X = df[trace_cols].T.values
    y = [website_name] * X.shape[0]

    features.extend(X)
    labels.extend(y)

# Convert to arrays
X = np.array(features)
y = np.array(labels)
print(X.shape)
print(y.shape)
print(X)
print(y)

# Normalize features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Reshape for Conv1D: (samples, time steps, features)
X = X.reshape(X.shape[0], X.shape[1], 1)

# Encode labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)
y_categorical = to_categorical(y_encoded)

# Split into train/test
X_train, X_test, y_train, y_test = train_test_split(X, y_categorical, test_size=0.2, random_state=42)
```

```python
trace_sets = {}
for f in csv_files:
    name = os.path.basename(f).split('_power_traces.csv')[0]
    df = pd.read_csv(f)
    trace_cols = [col for col in df.columns if col.startswith('Trace_')]
    traces = df[trace_cols].to_numpy().T
    trace_sets[name] = traces
    print(f"Loaded {name}: {traces.shape}")

# TVLA (pairwise t-test)
print("\nComputing pairwise TVLA...")

def compute_tvla(traces1, traces2):
    t_stats, _ = ttest_ind(traces1, traces2, axis=0, equal_var=False)
    return t_stats

tvla_results = {}
pair_names = []

for (name1, traces1), (name2, traces2) in combinations(trace_sets.items(), 2):
    t = compute_tvla(traces1, traces2)
    key = f"{name1} vs {name2}"
    tvla_results[key] = t
    pair_names.append(key)

    # 📊 Plot TVLA
    plt.figure(figsize=(12, 4))
    plt.plot(t, label=f'TVLA: {key}')
    plt.axhline(y=4.5, color='r', linestyle='--', label='Threshold (+4.5)')
    plt.axhline(y=-4.5, color='r', linestyle='--', label='Threshold (-4.5)')
    plt.title(f'TVLA {key}')
    plt.xlabel('Time Samples')
    plt.ylabel('t-value')
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()

# 📊 NICV Calculation
print("\nComputing NICV across all classes...")

all_traces = []
all_labels = []

for label, (name, traces) in enumerate(trace_sets.items()):
    all_traces.append(traces)
    all_labels.extend([label] * traces.shape[0])

X = np.vstack(all_traces)
y = np.array(all_labels)

# ➗ Compute NICV
def compute_nicv(X, y):
    nicv = []
    for i in range(X.shape[1]):
        total_var = np.var(X[:, i])
        group_means = [np.mean(X[y == c, i]) for c in np.unique(y)]
        between_var = np.var(group_means)
        nicv_val = between_var / total_var if total_var > 0 else 0
        nicv.append(nicv_val)
    return np.array(nicv)

nicv_vals = compute_nicv(X, y)

# 📊 Plot NICV
plt.figure(figsize=(12, 4))
plt.plot(nicv_vals, label='NICV')
plt.title('Normalized Inter-Class Variance (NICV)')
plt.xlabel('Time Samples')
plt.ylabel('NICV')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```

```python
model = Sequential([
    Conv1D(64, kernel_size=5, activation='relu', input_shape=(X.shape[1], 1)),
    MaxPooling1D(pool_size=2),
    Dropout(0.3),

    Conv1D(128, kernel_size=3, activation='relu'),
    MaxPooling1D(pool_size=2),
    Dropout(0.4),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(y_categorical.shape[1], activation='softmax')
])

# Compile
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train
print("\nTraining CNN model...")
history = model.fit(X_train, y_train, epochs=75, batch_size=32, validation_split=0.1)

# Evaluate
print("\nEvaluating on test set...")
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy * 100:.2f}%")

# Predictions
y_pred = model.predict(X_test)
y_pred_labels = np.argmax(y_pred, axis=1)
y_true_labels = np.argmax(y_test, axis=1)

# Classification report
print("\nClassification Report:")
print(classification_report(y_true_labels, y_pred_labels, target_names=le.classes_))

# Confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(y_true_labels, y_pred_labels))

# Save model and label encoder
model.save('cnn_model.h5')
import joblib
joblib.dump(le, 'label_encoder.pkl')
```

```python
num_traces_per_class = 2000

# Initialize confusion-like matrix
num_classes = len(le.classes_)
confusion_like_matrix = np.zeros((num_classes, num_classes), dtype=int)

print("\nBuilding confusion-like matrix for each website...")

for csv_file in csv_files:
    website_name = os.path.basename(csv_file).split('_power_traces.csv')[0]

    if website_name not in le.classes_:
        print(f"Skipping {website_name}: not found in label encoder classes.")
        continue

    # Read and extract trace data
    df = pd.read_csv(csv_file)
    trace_cols = [col for col in df.columns if col.startswith("Trace_")]
    X_site = df[trace_cols].T.values  # Each column is a trace
    X_site = X_site[:num_traces_per_class].astype(float)  # Limit to 2000

    if X_site.shape[0] < num_traces_per_class:
        print(f"Warning: only {X_site.shape[0]} traces available for {website_name}")

    # Normalize and reshape
    X_site = scaler.transform(X_site)
    X_site = X_site.reshape(X_site.shape[0], X_site.shape[1], 1)

    # Predict
    y_site_pred_probs = model.predict(X_site)
    y_site_pred = np.argmax(y_site_pred_probs, axis=1)

    # Count predictions
    pred_counts = Counter(y_site_pred)
    row_idx = le.transform([website_name])[0]
    for col_idx in range(num_classes):
        confusion_like_matrix[row_idx, col_idx] = pred_counts.get(col_idx, 0)

print("\nConfusion-like matrix (rows: true class, columns: predicted class):")
df_matrix = pd.DataFrame(confusion_like_matrix, index=le.classes_, columns=le.classes_)
print(df_matrix)

df_matrix.to_csv('confusion_like_matrix_7x7.csv')

plt.figure(figsize=(10, 8))
sns.heatmap(df_matrix, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion-like Matrix for 2000 Traces per Website")
plt.xlabel("Predicted Class")
plt.ylabel("True Class")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Parameters
num_trials = 50  # how many times to repeat the random sampling
trace_counts = list(range(1, 21))  # number of traces from 1 to 20
avg_ranks_per_class = {class_name: [] for class_name in le.classes_}

print("\nComputing rank vs number of traces...")

for num_traces in tqdm(trace_counts):
    ranks = []

    for website in le.classes_:
        # Read corresponding CSV
        csv_file = os.path.join(data_folder, f"{website}_power_traces.csv")
        df = pd.read_csv(csv_file)
        trace_cols = [col for col in df.columns if col.startswith("Trace_")]
        X_website = df[trace_cols].T.values.astype(float)

        if len(X_website) < num_traces:
            continue  # skip if not enough traces

        ranks_for_trials = []

        for _ in range(num_trials):
            selected = X_website[np.random.choice(len(X_website), num_traces, replace=False)]

            # Normalize and reshape
            selected_scaled = scaler.transform(selected)
            selected_scaled = selected_scaled.reshape(selected_scaled.shape[0], selected_scaled.shape[1], 1)

            # Predict and average probabilities
            probs = model.predict(selected_scaled)
            avg_probs = np.mean(probs, axis=0)

            # Get rank of true class
            sorted_indices = np.argsort(avg_probs)[::-1]  # descending
            true_class_index = le.transform([website])[0]
            rank = np.where(sorted_indices == true_class_index)[0][0] + 1  # +1 for 1-based rank
            ranks_for_trials.append(rank)

        # Store average rank for this website and trace count
        avg_rank = np.mean(ranks_for_trials)
        avg_ranks_per_class[website].append(avg_rank)

# Plotting
plt.figure(figsize=(12, 7))

for website, ranks in avg_ranks_per_class.items():
    plt.plot(trace_counts, ranks, label=website)

plt.xlabel('Number of Traces')
plt.ylabel('Average Rank of True Class')
plt.title('Rank Function vs Number of Traces')
plt.grid(True)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Dropout, Flatten, Dense
from sklearn.preprocessing import StandardScaler
import os
import pandas as pd

def compute_rank(scores, correct_key_index):
    """
    Compute the rank of the correct key.
    """
    sorted_indices = np.argsort(scores)[::-1]
    rank = np.where(sorted_indices == correct_key_index)[0][0]
    return rank

def rank_attack(model, X_test, y_test_encoded, n_values, le, repetitions=10):
    """
    Perform repeated attacks to estimate average rank evolution over increasing number of traces.

    Parameters:
    - model: Trained Keras model.
    - X_test: Test set features.
    - y_test_encoded: Encoded test labels.
    - n_values: List of n values (number of traces used in rank computation).
    - le: LabelEncoder used to encode/decode class labels.
    - repetitions: Number of repetitions for averaging rank.

    Returns:
    - ranks_mean: Mean rank per `n` over `repetitions`.
    """
    ranks_all = np.zeros((repetitions, len(n_values)))

    for rep in range(repetitions):
        # Shuffle test data
        idx = np.random.permutation(len(X_test))
        X_test_shuffled = X_test[idx]
        y_test_shuffled = y_test_encoded[idx]

        for i, n in enumerate(n_values):
            X_batch = X_test_shuffled[:n]
            y_batch = y_test_shuffled[:n]
            preds = model.predict(X_batch, verbose=0)  # shape: (n, num_classes)
            scores = np.sum(np.log(preds + 1e-40), axis=0)  # shape: (num_classes,)

            # Determine the correct key (most common in batch)
            correct_key_index = np.bincount(y_batch).argmax()
            rank = compute_rank(scores, correct_key_index)
            ranks_all[rep, i] = rank

    ranks_mean = np.mean(ranks_all, axis=0)
    return ranks_mean
```
